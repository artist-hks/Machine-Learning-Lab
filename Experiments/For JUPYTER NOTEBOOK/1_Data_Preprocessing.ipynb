## 📦 Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## ❗ Checking Missing Values

```python
df.isnull().sum()
```

## 🧹 Handling Missing Values (Drop Method)

```python
df = df.dropna()
df.isnull().sum()
```

## 🩹 Handling Missing Values using SimpleImputer
Missing numeric values are replaced using the mean of each column.

```python
from sklearn.impute import SimpleImputer

num_cols = df.select_dtypes(include=np.number)

imputer = SimpleImputer(strategy='mean')

df[num_cols.columns] = imputer.fit_transform(num_cols)

df.isnull().sum()
```

## 📊 Statistical Summary

```python
df.describe()
```

## 📈 Data Visualization - Histogram

```python
df.hist(figsize=(10,8))
plt.show()
```

## 🌊 Density Plot

```python
df.plot(kind='density', subplots=True, layout=(3,3), sharex=False, figsize=(10,8))
plt.show()
```

## 🎯 Feature Selection
Separate features (X) and target (y) from numeric columns.

```python
X = df.select_dtypes(include=[np.number]).iloc[:, :-1]
y = df.select_dtypes(include=[np.number]).iloc[:, -1]

X, y
```
