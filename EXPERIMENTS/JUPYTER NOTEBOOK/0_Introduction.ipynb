## 📘 Introduction to Machine Learning

Machine Learning (ML) is a field of Artificial Intelligence that allows computers to learn from data and make predictions.

ML systems improve performance automatically through experience without being explicitly programmed.

## 🧠 Types of Machine Learning

**1. Supervised Learning**  
- Uses labeled data  
- Example: Regression, Classification  

**2. Unsupervised Learning**  
- Uses unlabeled data  
- Example: Clustering  

**3. Reinforcement Learning**  
- Learns through rewards and punishments  

## 📓 Jupyter Notebook

Jupyter Notebook is an open-source tool used for data science and ML.

### Features:
- Code execution in cells  
- Visualization support  
- Easy documentation using Markdown  

## ☁️ Google Colab

Google Colab is a cloud-based platform to run Python notebooks.

### Advantages:
- No installation required  
- Free GPU support  
- Easy sharing  

## ❗ Common Errors in ML Lab

**1. FileNotFoundError**  
- Dataset not uploaded or wrong filename  

**2. ModuleNotFoundError**  
- Library not installed  

**3. ValueError**  
- Wrong data type or missing values  

**4. NameError**  
- Variable used before defining  

**5. IndentationError**  
- Incorrect spacing in code  

**6. SyntaxError**  
- Missing brackets or colons
